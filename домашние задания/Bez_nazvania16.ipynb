# 1 выгрузить данные 

```python
df=pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/1Customers.csv')
df
```

```python
df.rename(columns={"Annual Income ($)": "Income","Spending Score (1-100)": "Score"}, inplace=True)
df
```

# 2 Спецификация

$$
Score = \beta_0 + \beta_1*Income+\beta_2*Gender+\beta_3*Age+\beta_4*Profession
$$

```python
zaika = smf.ols(data=df,formula='Score ~ Income+ C(Gender)+Age+Profession'). fit()
zaika.params.round(3)
```

```python
df1 = df.head(3)
df1
```

```python
model1 = smf.ols('Score ~ CustomerID + Income +C(Profession)',data = df).fit()
```

```python
model1_pre = model1.predict(exog=df1, transform=True)
model1_pre.round(2).to_frame(name='Прогноз')
```
